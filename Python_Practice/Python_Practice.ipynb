# ++++++++++++Columns Transformer and HOTlablel encoder++++++++++++

```python
df = pd.read_csv(r'E:\notes\noorain_python\Python Class(09-Aug)\maati_ML\Simple and Miulti Linear Regression\50_Startups.csv')
df.head()
```

```python
X = df[['R&D Spend','Administration','Marketing Spend','State']]
y = df['Profit']
```

```python
X
```

# Encoding Categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [3])], remainder='passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
Ran = RandomForestRegressor()
```

```python
Ran.fit(X_train,y_train)
```

```python
print(Ran.predict(X_test))
print(np.array(y_test))
```

```python
Ran.score(X_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
```

```python
lin.fit(X_train,y_train)
```

```python
lin.score(X_test,y_test)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear')
```

```python
svr.fit(X_train,y_train)
```

```python
svr.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(RandomForestRegressor(), X_train,y_train, cv=4)
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
grid = GridSearchCV(RandomForestRegressor(),
                   {
                       'n_estimators' : [50,100,200]
                   }, return_train_score=False, cv = 4
                   )
```

```python
grid.fit(X_train, y_train)
```

```python
grid.cv_results_
```

```python
df1 =pd.DataFrame(grid.cv_results_)
```

```python
df1
```

```python
df2 = df1[['param_n_estimators','mean_test_score']]
```

```python
df2.plot(kind='bar')
```
