```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel(r'E:\notes\noorain_python\Python Class(19-Apr)\Project Data.xlsx')
data
```

```python
data.info()
```

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
```

```python
data.drop(['Year','Month'], axis=1, inplace=True)
```

```python
df = data.rename({'Month Number':'Month No.', 'Month Name':'Month', 2017:'Year' }, axis=1)
```

```python
df.columns
```

```python
df1 = data[['Segment','Country','Product',' Sales','Month']].groupby(['Country','Month'])[' Sales'].sum()
df = df1.sort_values(ascending=False).nlargest(n=10)
```

```python
plt.figure(figsize=(20,10))
df.plot(kind='bar')
```
