# 1. Which Business units are underperforming in 2018 and what are the key drivers?

```python
df5 = df4[df4['Year']==2018]
df6 = df5[['Group','Revenue']].groupby('Group')['Revenue'].agg('sum')
df7 = df6.sort_values(ascending=True)
```

```python
plt.figure(figsize=(10,5))
df7.plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Which Business units are underperforming in 2018 and what are the key drivers?')
```

# 2. How a group is performing vs previous year/period?

```python
df5 = pd.crosstab(df4['Group'], df4['Year'], values=df4['Revenue'], aggfunc=sum)
df6 = df5[[2017.0,2018.0]]
df6
```

```python
plt.figure(figsize=(10,20))
df6.plot(kind='bar')
plt.title('How a group is performing vs previous year/period?')
```

# 3. Where are the biggest variances from forecasts? And what recommendations do you have?

# 4. Analyze sales by top-10 performers and underperformers by country and product?

```python
df5 = df4[['CountryID','Revenue']].groupby('CountryID').sum()
df5
df6 = df5['Revenue'].nlargest(n=10)
df6
```

```python
df6.plot(kind='bar')
plt.title('sales by top-10 performers Countries')
```

```python
df5 = df4[['CountryID','Revenue']].groupby('CountryID').sum()
df5
df6 = df5['Revenue'].nsmallest(n=10)
df6
```

```python
df6.plot(kind='bar')
plt.title('sales by Lowest-10 performers Countries')
```

```python
df5 = df4[['Product','Revenue']].groupby('Product').sum()
df5
df6 = df5['Revenue'].nlargest(n=10)
df6
```

```python
df6.plot(kind='barh')
plt.title('sales by Top-10 performers Product')
```

```python
df5 = df4[['Product','Revenue']].groupby('Product').sum()
df5
df6 = df5['Revenue'].nsmallest(n=10)
df6
```

```python
df6.plot(kind='barh')
plt.title('sales by Lowest-10 performers Product')
```

# What other key insights can be drawn from the data?
